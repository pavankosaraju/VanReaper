```python
import requests
import re
from bs4 import BeautifulSoup
import csv
import random
import time
import numpy as np
```

```python
def get_postal_codes():
    
    area = []
    aname = []
    l = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_V"
    r = requests.get(l)
    s = BeautifulSoup(r.content,'html.parser')
    dt = s.find_all("td",limit = 180)
    for a in dt:
        area.append(a.find("b").text)
        names = a.find_all("a")
        n = ""
        for na in names:
            n = n + na.text+", "
        n = n.strip(" ,")
        aname.append(n)
    return area,aname
```

```python
pc,name = get_postal_codes()
```

```python
dt = dt[dt['YEAR']>2008]
```

```python
dt = dt.assign(Postal="",Areaname="")
```

```python
dt[dt['TYPE'] == '' or '' or '']
```

```python
for i in dt.index:
    st = str(dt.loc[i,'NEIGHBOURHOOD'])
    l = [x for x in range(len(name)) if st.lower() in name[x].lower()]
    if len(l)>0:
        dt.loc[i,'Postal'] = pc[l[0]]
        dt.loc[i,'Areaname'] = name[l[0]]
```

```python
dt.to_csv()
```
